```python
import pandas as pd

df = pd.read_csv("datasets/adult/adult.csv")
df.head()
```

```python
# Find and display duplicated rows grouped together
duplicated_mask = df.duplicated(keep=False)
duplicated_df = df[duplicated_mask].copy()

if len(duplicated_df) > 0:
    # Sort by all columns so identical rows appear together
    duplicated_df_sorted = duplicated_df.sort_values(by=list(df.columns)).reset_index()
    print(f"Found {df.duplicated().sum()} duplicate rows (excluding first occurrence)")
    print(f"Total rows involved in duplication: {len(duplicated_df)}\n")
    print("Duplicated rows (sorted so duplicates appear together):")
    print(duplicated_df_sorted)
else:
    print("No duplicates found in the dataset.")
```

```python
# Count how many duplicates exist
print(f"Number of duplicate rows: {df.duplicated().sum()}")

# Show the duplicated rows with their index
duplicated_rows = df[df.duplicated(keep=False)]
print(f"Total rows involved in duplication: {len(duplicated_rows)}")
duplicated_rows
```
